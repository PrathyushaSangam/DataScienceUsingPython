### About the data
The dataset is a csv file of 10,000 rows. Each row contains information about a pixel in a photograph of a map. There are 30 columns, with the first 29 being features about the pixel and the last being a target value, representing the "land use".

Columns indexed 0 and 1 represent the (x,y) co-ordinates of a pixel, indices 2-28 contain RGB values of a 3x3 grid, of the pixel (x,y) and its 8 neighbouring pixels.

The target labels are:
* 0 building
* 1 road
* 2 park or field
* 3 forest
* 4 water (river, sea or lake)

  Classify 0-Building vs all classes, extend to multiclass classification.

### <a name="step4"></a> Step 4: Decisions related to data cleaning and choice of features

**1\. Importance of color features:** From data visualization, we learn that, to determine the label of (x,y) pixel, the RGB values of only that pixel are insufficient, we need to know the surrounding colors as well. (for example building grid has some shades of blue, we can't say it is water). **Decision: We need all RGBs of all 9 pixels to be fed to the algorithm.**

**2\. Importance of location feature:** From the scatterplot we know that roads, forest(some parts), parks and roads are scattered on the map, whereas, buildings and majority of water have a specific location - i.e. by its (x,y) co-ordinates we can guess the probable land use. (this assumption is made expecting that in future, the classifiers will be used only for this map, if a new map is to be fed, location mustn't be considered). **Decision: Keep the location feature.**

**3\. Scaling the data:** The RGB values have similar scale and lie between 0-255, it makes no specific difference to scale them between 0-1. However, location co-ordinates have a different scale. Since some ML algorithms are sensitive to scaling it is better to scale all features by scaling them between 0-1. **Decision: use sci-kit learn's StandardScaler**


### <a name="step5"></a>Step 5: Stratified Train-Test Splits
Before performing the classification task, we will make Train Test splits in 80:20 ratio (industry recommended).
Since there are huge variations in number of samples of each category of land use, we can perform stratified sampling.

#### Naming convention of the train-test splits:
* X_train: Training data frame containing features - without labels
* y_train: Data labels for training
* X_test: testing data frame containing features - without labels
* y_test: Original data labels of test data

```python
print("Original data shape:",dataframe.shape)

X = dataframe.iloc[:,0:29]
print("Shape of X , without the land use labels:",X.shape)
#X.head()

y = dataframe.iloc[:,29]
print("Shape of y , land use labels:",y.shape)
#y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    stratify=y, 
                                                    test_size=0.2, random_state = 42)

# To view how the dataframes look like - Uncomment the below lines
#X_train.head()
#X_test.head()
#y_train.head()
#y_test.head()
```

```python
print("Checking the ratio of labels in test and train splits: We have same ratio of each label in train and test")

y_train.value_counts() / len(y_train)

y_test.value_counts() / len(y_test)
```

### <a name="step6"></a>Step 6: Building a one-vs-all (binary classifier)
The objective of this classifier is to identify buildings (label 0) among other labels.

#### Pre-processing 
* To support this functionality, we create a new boolean feature, which marks all samples of 0-Building as true and the rest as false. We perform the same action on test and train labels.
* We scale "X" to handle algorithms which are sensitive to scaling

#### Classification algorithms used for one-vs-all classification
* SGD classifier
* Logistic regression
* SVM

#### Performance measures
* Cross validation accuracy scores on training set
* Accuracy, Precision and Recall on Training set
* Accuracy, Precision and Recall on Test set (validation)
* ROC curves and AUC values for comparision of methods

```python
# Modifying "y" for binary classifier 0-building vs all

y_train_0 = (y_train == 0)
y_test_0 = (y_test == 0)
```

```python
# Feature scaling using Standard scaler for "X"

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train.astype(np.float64))
X_test = scaler.fit_transform(X_test.astype(np.float64))
```

### <a name="sgd"></a>SGD classifier

```python
#  Fitting SGD classifier

sgd_clf = SGDClassifier(max_iter=5, random_state=42)
sgd_clf.fit(X_train, y_train_0)

# Cross Val Score - Training
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_0, cv=3, scoring="accuracy")

from sklearn.metrics import precision_score, recall_score, accuracy_score

# Performance measures on Training set
sgd_train_pred = sgd_clf.predict(X_train)

sgd_train_score = accuracy_score(y_train_0,sgd_train_pred)
sgd_train_precision = precision_score(y_train_0,sgd_train_pred)
sgd_train_recall = recall_score(y_train_0,sgd_train_pred)


# Performance measures on Test set

sgd_test_pred = sgd_clf.predict(X_test)

sgd_test_score = accuracy_score(y_test_0,sgd_test_pred)
sgd_test_precision = precision_score(y_test_0,sgd_test_pred)
sgd_test_recall = recall_score(y_test_0,sgd_test_pred)


# Model Evaluation

sgd_model = pd.DataFrame({
    'Performance Measures SGD': ['Training Accuracy','Training Precision', 'Training Recall', 'Test Accuracy','Test Precision'
                                     ,'Test Recall'],
    'Scores': [sgd_train_score,sgd_train_precision,sgd_train_recall,sgd_test_score,sgd_test_precision,sgd_test_recall]})

sgd_model
```

### <a name="log"></a>Logistic Regression

```python
#logistic regression

logreg = LogisticRegression(random_state=42)
logreg.fit(X_train, y_train_0)

# Cross Val Score - Training
cross_val_score(logreg, X_train, y_train_0, cv=3, scoring="accuracy")

# Performance measures on Training set
log_train_pred = logreg.predict(X_train)

logreg_train_score = accuracy_score(y_train_0,log_train_pred)
logreg_train_precision = precision_score(y_train_0,log_train_pred)
logreg_train_recall = recall_score(y_train_0,log_train_pred)


# Performance measures on Test set

log_test_pred = logreg.predict(X_test)

logreg_test_score = accuracy_score(y_test_0,log_test_pred)
logreg_test_precision = precision_score(y_test_0,log_test_pred)
logreg_test_recall = recall_score(y_test_0,log_test_pred)


# Model Evaluation

log_model = pd.DataFrame({
    'Performance Measures LogReg': ['Training Accuracy','Training Precision', 'Training Recall', 'Test Accuracy', 'Test Precision'
                                     ,'Test Recall'],
    'Scores': [logreg_train_score,logreg_train_precision,logreg_train_recall,logreg_test_score,logreg_test_precision,logreg_test_recall]})

log_model
```

### <a name="svm"></a>Support Vector Machines

```python
# Support Vector Machines

svc = SVC(random_state=42)
svc.fit(X_train, y_train_0)

# Cross Val Score - Training
cross_val_score(svc, X_train, y_train_0, cv=3, scoring="accuracy")

# Performance measures on Training set
svc_train_pred = svc.predict(X_train)

svc_train_score = accuracy_score(y_train_0,svc_train_pred)
svc_train_precision = precision_score(y_train_0,svc_train_pred)
svc_train_recall = recall_score(y_train_0,svc_train_pred)


# Performance measures on Test set

svc_test_pred = svc.predict(X_test)

svc_test_score = accuracy_score(y_test_0,svc_test_pred)
svc_test_precision = precision_score(y_test_0,svc_test_pred)
svc_test_recall = recall_score(y_test_0,svc_test_pred)


# Model Evaluation

svc_model = pd.DataFrame({
    'Performance Measures SVM': ['Training Accuracy','Training Precision', 'Training Recall', 'Test Accuracy', 'Test Precision'
                                     ,'Test Recall'],
    'Scores': [svc_train_score,svc_train_precision,svc_train_recall,svc_test_score,svc_test_precision,svc_test_recall]})

svc_model
```

```python
## ROC curves function taken from textbook ##
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)


#ROC for SGD, SVC and Log Reg

y_scores_sgd = cross_val_predict(sgd_clf, X_train, y_train_0, cv=3,
                             method="decision_function")
y_scores_svc = cross_val_predict(svc, X_train, y_train_0, cv=3,
                             method="decision_function")

y_scores_log = cross_val_predict(logreg, X_train, y_train_0, cv=3,
                             method="decision_function")



from sklearn.metrics import roc_curve,roc_auc_score

fpr_sgd, tpr_sgd, thresholds_sgd = roc_curve(y_train_0,y_scores_sgd)
fpr_svc, tpr_svc, thresholds_svc = roc_curve(y_train_0,y_scores_svc)
fpr_log, tpr_log, thresholds_log = roc_curve(y_train_0,y_scores_svc)


plt.figure(figsize=(8, 6))
plt.plot(fpr_sgd, tpr_sgd, "b:", linewidth=2, label="SGD")
plot_roc_curve(fpr_svc, tpr_svc, "SVM")
plot_roc_curve(fpr_log, tpr_log, "LOG Reg")

plt.legend(loc="lower right", fontsize=16)
plt.show()


print('AUC of SGD:', roc_auc_score(y_train_0, y_scores_sgd))
print('AUC of Log Reg:',roc_auc_score(y_train_0, y_scores_log))
print('AUC of SVM:',roc_auc_score(y_train_0, y_scores_svc))
```

#### <a name="findings6"></a>  Performance Evaluation and Findings of Binary Classification algorithms (one vs all)

1\. On the whole, SVM performed the best, followed by logistic regression and lastly SGD. This can be seen in individual performance scores - accuracy, precision and recall during testing and training phases, as well as from the ROC curves.

2\. Note: The SVM and Log Reg curves are overlapping due to which one of them seems missing in the ROC graph.

3\. None of the algorithms showed any explicit signs of overfit or underfit, since they exhibited similar performance during train and test phases, this is probably due to the simplicity of data and StandardScaling.

### <a name="step7"></a> Step 7: Multiclass Classification

The objective of this classifier is to classify data under one of labels
* 0 building
* 1 road
* 2 park or field
* 3 forest
* 4 water (river, sea or lake)

#### Pre-processing 
* We can keep the original labels (y-values)
* We use the scaled "X" values from the previous step, to use algorithms which are sensitive to scaling.

#### Algorithms used
* Random Forest Classifier
* K-neighbours Classifier

#### Performance measures
* Cross validation accuracy scores on training set
* Accuracy on Training and Test sets
* Confusion Matrix

### <a name="rft"></a>Random Forest Classifier

```python
# Random Forest Trees

rft = RandomForestClassifier(n_jobs=2, random_state=42)
rft.fit(X_train, y_train)

# Cross Val Score - Training
cross_val_score(rft, X_train, y_train, cv=3, scoring="accuracy")

# Performance measures on Training set
rft_train_pred = rft.predict(X_train)
rft_train_score = rft.score(X_train, y_train)


# Performance measures on Test set
rft_test_pred = rft.predict(X_test)
rft_test_score = rft.score(X_test, y_test_pred)
```

### <a name="knn"></a>K-Neighbours Classifier

```python
np.random.seed(42)

#KNN k=3

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)

#Cross-val Score
cross_val_score(knn, X_train, y_train, cv=3, scoring="accuracy")

# Training 
knn_train_pred = knn.predict(X_train)
knn_train_score = knn.score(X_train, y_train)


# Testing
knn_test_pred = knn.predict(X_test)
knn_test_score = knn.score(X_test, y_test)
```

### <a name="findings7"></a>Multi-class classification evaluation: Visualizing the confusion matrices

```python
multi_model = pd.DataFrame({
    'Performance Measures': ['Training Accuracy of RFT', 'Testing Accuracy of RFT','Training Accuracy of KNN','Testing Accuracy of KNN'],
    'Scores': [rft_train_score,rft_test_score,knn_train_score,knn_test_score]})

multi_model
```

```python
## Plotting function Taken from the text book ##
def plot_confusion_matrix(matrix):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111)
    cax = ax.matshow(matrix)
    fig.colorbar(cax)


conf_mx_rft = confusion_matrix(y_train, rft_train_pred)
conf_mx_rft
   
plt.matshow(conf_mx_rft, cmap='Purples')
plt.show()


#one 2-park is labelled as 3-forest, twelve 1-road as 0-building, all the others were categorized perfectly
```

```python
## Plotting function Taken from the text book ##
conf_mx_knn = confusion_matrix(y_train, knn_train_pred)
conf_mx_knn
    
plt.matshow(conf_mx_knn, cmap='Greens')
plt.show()

#40 2-park recognized as 3-forest, 31 1-road recognized as 0-building, 59 1-road recognized as 0-building
#14 3-forest recognized as 2-park
```

### Findings of multiclass classification
1\. Random Forest Trees performed exceptionally well with 99.8 accuracy in training and during test phase with ~100% accuracy! which makes one wonder this possibility. The wrongly identified labels are 1 "2-park" which was classified as "3-forest" and 12 1-road as 0-building, all the others were categorized perfectly.

2\. K-neighbour classifier has a slightly lower performance than RFT, but hit accuracy as high as 97.9 training and 96% in testing phases. The wrongly identified classes were
40 2-park recognized as 3-forest, 31 0-building recognized as 1-roads, 59 1-roads recognized as 0-building, 14 3-forest recognized as 2-park.
